```python
from qiskit import *
from qiskit.tools.visualization import *
from qiskit.tools.monitor import *
from matplotlib import *
```

```python
qr = QuantumRegister( 2 )   # 2 quantum registers
cr = ClassicalRegister( 2 )   # 2 classical registers to measure 2 qubits
```

```python
circ = QuantumCircuit( qr, cr )
circ.draw()
```

```python
circ.h( qr[0] )
circ.cx( qr[0], qr[1] )
circ.measure( qr, cr )
```

```python
circ.draw( output = "mpl" )
```

```python
sim = Aer.get_backend( "qasm_simulator" )   # using a simulator
```

```python
res = execute( circ, sim ).result()
```

```python
plot_histogram( res.get_counts( circ ) )
```

```python
# remember to do IBMQ.load_account()
provider = IBMQ.get_provider( "ibm-q" )
```

```python
qcomp = provider.get_backend( "ibmq_16_melbourne" )
```

```python
job = execute( circ, qcomp )
```

```python
job_monitor( job )
```

```python
res = job.result()   # executed on a real quantum computer
```

```python
plot_histogram( res.get_counts( circ ) )
```
